# Exploring OpenCV

### Notebook Imports

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

### Sanity check: Testing if OpenCV can read and open images

```python
brown_image_path = './test_images/brown.png'
img = cv2.imread(brown_image_path)
cv2.imshow('brown image', img)
# Ending the code
cv2.waitKey(0) # waits until the x is pressed
cv2.destroyAllWindows() # destroys all windows
```

### Displaying the image -> Both match

```python
brown_image = plt.imread(brown_image_path)
plt.imshow(brown_image)
plt.axis('off')
plt.show()
print('Shape:', brown_image.shape)
```

### Exploring Color Spaces -> HSV (Hue Saturation Value) is recommended

```python
hsv_img = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
cv2.imshow('HSV of brown image', hsv_img)
# Ending the code
cv2.waitKey(0) # waits until the x is pressed
cv2.destroyAllWindows() # destroys all windows
```
