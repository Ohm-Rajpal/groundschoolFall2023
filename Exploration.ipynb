### Trying to split the color of an RGB image

### Example of arithmetic operators on binary images

```python
half_background = cv2.imread('./Exploration Images/BlackWhite.png')
circle = cv2.imread('./Exploration Images/CircleCircle.png')

cv2.imshow('original background', half_background)
cv2.waitKey(0)

cv2.imshow('original circle', circle)
cv2.waitKey(0)

# Remember that the direction does matter
# And is only true when both pixels in the images are greater than zero
and_operator = cv2.bitwise_and(circle, half_background, mask=None)
# Or is true when at least one pixel in the image is greater than zero (non-black color)
or_operator = cv2.bitwise_or(circle, half_background, mask=None)
# Not inverts off and on pixels. Takes only one img and a mask
not_operator = cv2.bitwise_not(circle)

cv2.imshow('and operator', and_operator)
cv2.waitKey(0)

cv2.imshow('or operator', or_operator)
cv2.waitKey(0)

cv2.imshow('not operator', not_operator)
cv2.waitKey(0)

cv2.destroyAllWindows()
```

### Example of resizing images

```python
# Reduce the circle image by half

# original circle
cv2.imshow('circle original', circle)
cv2.waitKey(0)

decrease_width = 100
decrease_height = 100
decrease = (decrease_width, decrease_height)

decreased_circle = cv2.resize(circle, decrease, interpolation=cv2.INTER_AREA)
cv2.imshow('decreased circle', decreased_circle)
cv2.waitKey(0)


increase_width = circle.shape[1] * 3
increase_height = circle.shape[0] * 3

increase = (increase_width, increase_height)

increased_circle = cv2.resize(circle, increase, interpolation=cv2.INTER_AREA)
cv2.imshow('increased_circle', increased_circle)
cv2.waitKey(0)
cv2.destroyAllWindows()
```

### How to crop images -> img[y1:y2, x1:x2]

```python
# Lets say we want to crop the flower image
# Order of cropping images
# [y1:y2, x1:x2]

# First display the flower
flower_img = cv2.imread('./Exploration Images/Flower.webp')
cv2.imshow('flower', flower_img)
cv2.waitKey(0)

cv2.imshow('flower', flower_img[75:300, 180:320])
cv2.waitKey(0)

cv2.destroyAllWindows()
```

### Annotating Images (Rectangle Example). Remember coordinates in OpenCV are like the normal xy plane but flipped over the x axis so x increases as you go right and y increases as you move down 

```python
dog_img = cv2.imread('./Exploration Images/HappyDoggo.jpg')

left_edge =(300,100)
right_edge = (530, 275)

cv2.rectangle(dog_img, left_edge, right_edge, (0,191,255), thickness= 2, lineType=cv2.LINE_8) 

show_img_custom(dog_img)
```

### Exploring Color Spaces

```python
bright_cube = cv2.imread('./Exploration Images/BrightCubeImg.png')
dark_cube = cv2.imread('./Exploration Images/DarkCubeImg.png')

show_img_custom(bright_cube)
show_img_custom(dark_cube)

bright_cube_HSV = cv2.cvtColor(bright_cube, cv2.COLOR_BGR2HSV)
dark_cube_HSV = cv2.cvtColor(dark_cube, cv2.COLOR_BGR2HSV)

show_img_custom(bright_cube_HSV)
show_img_custom(dark_cube_HSV)
```
